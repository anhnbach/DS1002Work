```python
import numpy as np
from PIL import Image
import requests

```

```python
print("Please input the url of an image")
url = input()
photo = np.asarray(Image.open(requests.get(url, stream=True).raw))
photo = photo.astype(np.uint8)

def red_photo(photo):
  photo_red = photo.copy()
  photo_red [:, :, 1] = 0
  photo_red [:, :, 2] = 0
  return photo_red

def green_photo(photo):
  photo_green = photo.copy()
  photo_green[:, :, 0] = 0
  photo_green[:, :, 2] = 0
  return photo_green

def blue_photo(photo):
  photo_blue = photo.copy()
  photo_blue[:, :, 0] = 0
  photo_blue[:, :, 1] = 0
  return photo_blue

def redandgreen_photo(photo):
  photo_rg = photo.copy()
  photo_rg = red_photo(photo) + green_photo(photo)
  return photo_rg

def redandblue_photo(photo):
  photo_rb = photo.copy()
  photo_rb = red_photo(photo) + blue_photo(photo)
  return photo_rb

def greenandblue_photo(photo):
  photo_gb = photo.copy()
  photo_gb = green_photo(photo) + blue_photo(photo)
  return photo_gb

def finalcheck(photo):
  photo_final = photo.copy()
  photo_final = red_photo(photo) + blue_photo(photo) + green_photo(photo)
  return photo_final

print("This is the original image")
display(photo)

print("This is the red-filtered version")
display(red_photo(photo)) #1

print("This is the green-filtered version")
display(green_photo(photo)) #2

print("This is the blue-filtered version")
display(blue_photo(photo)) #3

print("This is the version with both Red and Green values but no Blue")
display(redandgreen_photo(photo)) #4

print("This is the version with Red and Blue values but no Green")
display(redandblue_photo(photo)) #5

print("This is a version with Green and Blue values but no Red")
display(greenandblue_photo(photo)) #6


print("This is the original version recreated from combining all of the red-filtered, blue-filtered, and green-filtered versions")
display(finalcheck(photo))
final_photo = red_photo(photo) + blue_photo(photo) + green_photo(photo)
if(np.array_equal(photo[0,0], final_photo[0,0])):
  print("The photos are identical")
else:
  print("The photos are not identical")
```
